```python
import re
from bs4 import BeautifulSoup
import urllib.request
import lxml
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
webpage=urllib.request.urlopen("https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)").read()
```

```python
soup=BeautifulSoup(webpage,"html5lib")
```

```python
IMF=pd.DataFrame(columns=['Country','IMF'])
WorldBank=pd.DataFrame(columns=['Country','WorldBank'])
UN=pd.DataFrame(columns=['Country','UN'])
table_columns=[IMF,WorldBank,UN]
name=["IMF","WorldBank","UN"]
```

```python

tables=soup.findAll('table',{"class":"wikitable sortable"})
for i,table in enumerate(tables[0:3]):
    for tab in table.findAll('tr')[2:]:
        tab_raw=(tab.text).strip()
        country_re=re.findall("[A-Z].+",tab_raw)
        country=(country_re[0].split("[")[0])
        gdp_re=re.findall("[0-9].+$",tab_raw)[0]
        if "♠" in gdp_re:
            gdp=int((((gdp_re).replace(",", "")).split("[")[0]).split("♠")[1])
        else:
            gdp=int(((gdp_re).replace(",", "")).split("[")[0])
        table_columns[i]=table_columns[i].append({"Country":country,"{}".format(name[i]):gdp},ignore_index=True)
```

```python
for i in range(len(table_columns)):
    table_columns[i].set_index("Country",inplace=True)
```

```python
%matplotlib tk
df=pd.merge(table_columns[0],table_columns[1],how="inner",left_index=True,right_index=True)
df=pd.merge(df,table_columns[2],how="inner",left_index=True,right_index=True)
```

```python
df[:10].plot(kind="barh")
plt.xlabel("GDP (USD)")
```
